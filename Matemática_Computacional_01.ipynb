```python
# Método da bisseção com o uso do Teorema de Bolzano
import matplotlib.pyplot as plt
import numpy as np
f1 = lambda x: x**2-2 # Separação da função f(x) = x^2 + (log x) - 2 = 0 em duas funções
f2 = lambda x: -np.log(x)

x = np.linspace(0.01, 2, 100)
plt.plot(x, f1(x), x, f2(x))

plt.title('Localização de raízes por gráficos')
plt.legend(['$f_1(x)=x^2-2$','$f_2(x)=-ln(x)$'])
plt.ylabel('Eixo $y$')
plt.xlabel('Eixo $x$')
plt.grid()
```

```python
# Calculando os pontos médios entre dois pontos que contém a raiz da função e aplicando o Teorema de Bolzano
def bissec(f, a, b, eps=0.001):
    err = b-a
    x_ant = a
    while err > eps:
        x = (a+b)/2.0
        if f(x) == 0: # se o ponto médio for zero, pare
            break
        elif f(x)*f(a) < 0: # Aplicação do teorema de Bolzano
            b = x # o extremo superior do intervalo assume o valor de x
        else:
            a = x # o extremo inferior do intervalo assume o valor de x
        err = abs(x - x_ant)/abs(x)  # valor absoluto abs()
        x_ant = x
        print ("x = %.4f"%x,
               "E = %.4f"%err)

    return (x)

f = lambda x: x**2+np.log(x)-2
bissec(f,1.25,1.5,0.001) # poderia ter implementado diretamente f dentro da chamada de bissec()
```

```python
# uma função para calcular as diferenças divididas
def difdiv(x, y):
  m=len(x)
  dd=y.copy()
  for k in range(1,m):
    dd[k:m]=(dd[k:m] - dd[k-1])/(x[k:m]-x[k-1])

  return dd
```

```python
import numpy as np

# Método da Bisseção
def bissec(f, a, b, eps):
    err = b-a
    x_ant = a
    countB = 0 # Definindo uma variável para contar o número de iterações feitas
    while err > eps:
        x = (a+b)/2.0
        if f(x) == 0: # se o ponto médio for zero, pare
            break
        elif f(x)*f(a) < 0: # Aplicação do teorema de Bolzano
            b = x # o extremo superior do intervalo assume o valor de x
        else:
            a = x # o extremo inferior do intervalo assume o valor de x
        err = abs(x - x_ant)  # cálculo do erro absoluto
        countB += 1
        x_ant = x
        print ("t = %.15f"%x, "Erro = %.10f"%err)

    print(f"Iterações: {countB}")

f = lambda t: -2.5*(2*t**2 - np.cos(2*t)) # função da situação problema
bissec(f,0.4,0.6,0.000001)
print()
```
